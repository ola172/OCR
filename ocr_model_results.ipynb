```python
import easyocr
import pytesseract
import matplotlib.pyplot as plt
from PIL import Image
```

```python
image_paths = ["test_images/testocr.png",
               "test_images/distorted_text.jpg",
               "test_images/largepreview.png",
               "test_images/sample1.a36a230755dc.jpg",
               "test_images/Handwriting-test-dataset-for-OCR-operation.png",
               "test_images/mixed_font_2.png",
               "test_images/low_contrast.jpg",
               "test_images/computer-vision.jpg"]
```

## Tesseract OCR

```python
# Tesseract OCR
def run_tesseract(image_path):
    img = Image.open(image_path)
    return img, pytesseract.image_to_string(img)
```

```python
# best case image
image, text = (run_tesseract(image_paths[0]))
plt.imshow(image)
print(text)
```

```python
# distorted text
image, text = (run_tesseract(image_paths[1]))
plt.imshow(image)
print(text)
```

```python
# tiny text
image, text = (run_tesseract(image_paths[2]))
plt.imshow(image)
print(text)
```

```python
image, text = (run_tesseract(image_paths[3]))
plt.imshow(image)
print(text)
```

```python
# handwritten
image, text = (run_tesseract(image_paths[4]))
plt.imshow(image)
print(text)
```

```python
# Mixed fonts
image, text = (run_tesseract(image_paths[5]))
plt.imshow(image)
print(text)
```

```python
# low contrast
image, text = (run_tesseract(image_paths[6]))
plt.imshow(image)
print(text)
```

```python
# real photgraphed document 
image, text = (run_tesseract(image_paths[7]))
plt.imshow(image)
print(text)
```

## EasyOCR

```python
def run_easyocr(image_path):
    image = Image.open(image_path)
    reader = easyocr.Reader(['en'])
    return image,reader.readtext(image_path, detail = 0)
```

```python
# bestcase image
image, text = (run_easyocr(image_paths[0]))
plt.imshow(image)
print(text)
```

```python
# distorted text
image, text = (run_easyocr(image_paths[1]))
plt.imshow(image)
print(text)
```

```python
# tiny text
image, text = (run_easyocr(image_paths[2]))
plt.imshow(image)
print(" ".join(text))
```

```python
image, text = (run_easyocr(image_paths[3]))
plt.imshow(image)
print(" ".join(text))
```

```python
# handwritten
image, text = (run_easyocr(image_paths[4]))
plt.imshow(image)
print(text)
```

```python
# Mixed fonts
image, text = (run_easyocr(image_paths[5]))
plt.imshow(image)
print(" ".join(text))
```

```python
# low contrast
image, text = (run_easyocr(image_paths[6]))
plt.imshow(image)
print(" ".join(text))
```

```python
# real photgraphed document 
image, text = (run_easyocr(image_paths[7]))
plt.imshow(image)
print(" ".join(text))
```

## Conclusion

## in cases where image and text are clear, both models perform well.

## Tesseract is better than easyocr in case:
>> ### 1. real photgraphed document.
>> ### 2. handwritten.
>> ### 3. large paragraph (easyocr deop some words and somtimes add words not in image).
>> ### 4. Text divided into columns like researches.

## Easyocr is better than Tesseract in case:
>> ### 1. low contrast text image (we can make some preprocessing to enhance image before apply ocr.).
>> ### 2. mixed fonts text image ( easyocr is little better than tesseract but both is not reliable.).
